```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Add this import statement

# Assuming you have already defined jenis_bencana_tertinggi_per_kecamatan
jenis_bencana_tertinggi_per_kecamatan  = pd.read_excel('bencana jakarta (1).xlsx')
jenis_bencana_tertinggi_per_kecamatan = jenis_bencana_tertinggi_per_kecamatan.sort_values(by='jumlah_korban_bencana', ascending=False)


plt.figure(figsize=(12, 6))
sns.barplot(x='kecamatan', y='jumlah_korban_bencana', hue='jenis_bencana', data=jenis_bencana_tertinggi_per_kecamatan)
plt.title('Jumlah Korban Bencana Tertinggi per Jenis Bencana di Setiap Kecamatan (Diurutkan)')
plt.xlabel('Kecamatan')
plt.ylabel('Jumlah Korban Bencana')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Jenis Bencana', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Assuming your Excel file has columns 'jenis_bencana', 'tanggal_mulai', 'jam_kejadian', 'tanggal_selesai'

df['tanggal_mulai'] = pd.to_datetime(df['tanggal_mulai'])
df['jam_kejadian'] = pd.to_datetime(df['jam_kejadian'], format='%H:%M:%S').dt.time
df['tanggal_selesai'] = pd.to_datetime(df['tanggal_selesai'])

# Calculate duration in hours
df['duration_hours'] = (df['tanggal_selesai'] - df['tanggal_mulai']).dt.total_seconds() / 3600

# Sort DataFrame by start date
df.sort_values(by='tanggal_mulai', inplace=True)

# Group by disaster type
grouped_by_category = df.groupby('jenis_bencana')

# Plotting for each category
for category, group in grouped_by_category:
    plt.figure(figsize=(10, 5))

    # Plotting
    bars = plt.bar(group['tanggal_mulai'], group['duration_hours'], width=0.8, align='edge', color='skyblue', label=f'{category} Duration (hours)')

    # Add textual information above each bar
    for bar, duration in zip(bars, group['duration_hours']):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, f'{duration:.2f} hours', ha='center', va='bottom')

    plt.title(f'Time Series of {category} Duration')
    plt.xlabel('Start Date')
    plt.ylabel('Duration (hours)')
    plt.xticks(rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
